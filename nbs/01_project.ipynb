```python
#| export
class OVHProject():
    def __init__(self, id:str, region:str=None, consumption_mode:str=None):
        """
        OVH Project class, storing project information

        Parameters
        ----------
        id : str, required
            OVH project id
        region : str, optional
            Project ressources region
        """
        self.id = id
        self.base_url = f'/cloud/project/{self.id}'
        self.region = None

    def set_region(self, region:str):
        """
        Sets the project's region

        Parameters
        ----------
        region : str, required
            Project ressources region
        """
        self.region = region
```

Let's initiate a OVHProject

```python
#| eval: false
project = OVHProject('PROJECT_ID')
```

it has a base url

```python
#| eval: false
project.base_url
```

We can also set it's region to manage resources in a specific region

```python
#| eval: false
project.set_region('RBX')
```
